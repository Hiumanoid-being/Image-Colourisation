```python
import kagglehub
import shutil
from pathlib import Path

# Download dataset
path = kagglehub.dataset_download("benjaminkz/places365")
print("KaggleHub downloaded to:", path)

# Define your project data folder
project_data_path = Path.cwd() / "data" / "raw"
project_data_path.mkdir(parents=True, exist_ok=True)

# Copy downloaded files into your repo folder
for item in Path(path).iterdir():
    dest = project_data_path / item.name
    if item.is_dir():
        shutil.copytree(item, dest, dirs_exist_ok=True)
    else:
        shutil.copy(item, dest)

print(f"âœ… Dataset copied to: {project_data_path}")
```
